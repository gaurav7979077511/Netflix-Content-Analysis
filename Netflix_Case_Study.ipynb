### 3.1 Data Conversion

Data type conversion is a crucial step in data preprocessing. It ensures that each column is in an appropriate format, making data analysis more efficient and accurate. Incorrect data types can lead to unexpected behavior during computations, inefficient storage usage, and complications when performing operations such as filtering or grouping. Below is the table describing the necessary data type conversions:


| SL No. | **Column Name**   | **Current Data Type** | **Desired Data Type** | **Reason for Conversion** |
|--------|--------------------|------------------------|-----------------------|---------------------------|
| 1      | **`show_id`**      | `object`               | `string`              | ID values should remain as strings to prevent unwanted numerical operations. |
| 2      | **`type`**         | `object`               | `category`            | Limited unique values; converting to category reduces memory usage. |
| 3      | **`title`**        | `object`               | `string`              | Textual data should remain as a string for analysis purposes. |
| 4      | **`director`**     | `object`               | `string`              | Names of directors should remain strings, though missing values need handling. |
| 5      | **`cast`**         | `object`               | `string`              | Similar to `director`, names should remain strings; missing values require handling. |
| 6      | **`country`**      | `object`               | `category`            | Limited unique values; converting to category optimizes storage. |
| 7      | **`date_added`**   | `object`               | `datetime`            | Dates should be in datetime format for easy manipulation and analysis. |
| 8      | **`release_year`** | `int64`                | `int`                 | Year values are integers, so no change is required. |
| 9      | **`rating`**       | `object`               | `category`            | Limited unique values; converting to category improves efficiency. |
| 10     | **`duration`**     | `object`               | `int (minutes)`       | Converting to integer allows for numerical analysis on duration. |
| 11     | **`listed_in`**    | `object`               | `string`              | Genre categories should remain as strings. |
| 12     | **`description`**  | `object`               | `string`              | Descriptions should remain as textual data. |


By performing these conversions, the dataset becomes more structured, optimized for analysis, and easier to process in subsequent steps.



```python
# Convert 'show_id' to string to ensure it is treated as a unique identifier for each row
data['show_id'] = data['show_id'].astype('string')

# Convert 'type' to category for memory optimization and efficient analysis
data['type'] = data['type'].astype('category')

# Convert 'title' to string to preserve the textual data for analysis
data['title'] = data['title'].astype('string')

# Convert 'director' to string to ensure proper handling of textual data
data['director'] = data['director'].astype('string')

# Convert 'cast' to string to handle textual data about actors
data['cast'] = data['cast'].astype('string')

# Convert 'country' to category to optimize memory and enable efficient grouping
data['country'] = data['country'].astype('category')

# Convert 'date_added' to datetime format for time-based analysis
data['date_added'] = pd.to_datetime(data['date_added'].str.strip())

# Convert 'release_year' to integer for numerical computations and analysis
data['release_year'] = data['release_year'].astype('int')

# Convert 'rating' to category for efficient analysis of rating distribution
data['rating'] = data['rating'].astype('category')

# Extract numeric values from 'duration' and convert to integer; handle missing values appropriately
data['duration'] = data['duration'].apply(lambda x: int(x.split()[0]) if not pd.isnull(x) else x)

# Convert 'listed_in' to string to preserve genre information as textual data
data['listed_in'] = data['listed_in'].astype('string')

# Convert 'description' to string to retain summary information about the content
data['description'] = data['description'].astype('string')
```

### 3.2 Exploding Columns 💥  

In the dataset, certain columns like **cast**, **director**, **listed_in**, and **country** contain multiple values separated by commas. To perform detailed analysis, it is beneficial to explode these columns, where each value gets its own row.  

1. **Cast**:  
   - The **cast** column lists multiple actors in a single row.  
   - Exploding this column allows us to analyze the frequency and contributions of individual actors.  

2. **Director**:  
   - The **director** column contains multiple directors for some entries.  
   - By exploding this column, we can evaluate the contributions of each director separately.  

3. **Listed_in**:  
   - This column represents genres for movies and TV shows.  
   - Exploding it enables us to study which genres are most common in the Netflix catalog.  

4. **Country**:  
   - The **country** column lists multiple countries where some content was produced.  
   - Exploding it helps analyze the geographical distribution of content and identify trends across countries.  

This step is crucial for granular insights into the data and will help answer key questions, such as:  
- Which actors and directors contribute the most to Netflix content?  
- What genres dominate Netflix’s catalog?  
- How is content distributed across different countries?  

Exploding these columns allows for more focused and meaningful analysis. Let’s proceed with implementing this transformation! 🚀


```python
# Split the 'cast' column into a list of actors (by splitting on commas) and explode the list into separate rows
data = data.assign(cast=data.cast.str.split(',')).explode('cast')

# Reset the index after exploding and rename the 'cast' column to 'actor'
data = data.reset_index(drop=True).rename(columns={'cast': 'actor'})

# Remove any leading or trailing whitespace from the 'actor' column
data['actor'] = data['actor'].str.strip()

# Split the 'listed_in' column into a list of genres (by splitting on commas) and explode the list into separate rows
data = data.assign(listed_in=data.listed_in.str.split(',')).explode('listed_in')

# Reset the index after exploding and rename the 'listed_in' column to 'genre'
data = data.reset_index(drop=True).rename(columns={'listed_in': 'genre'})

# Remove any leading or trailing whitespace from the 'genre' column
data['genre'] = data['genre'].str.strip()

# Split the 'director' column into a list of directors (by splitting on commas) and explode the list into separate rows
data = data.assign(director=data.director.str.split(',')).explode('director')

# Remove any leading or trailing whitespace from the 'director' column
data['director'] = data['director'].str.strip()

# Split the 'country' column into a list of countries (by splitting on commas) and explode the list into separate rows
data = data.assign(country=data.country.str.split(',')).explode('country')

# Remove any leading or trailing whitespace from the 'country' column
data['country'] = data['country'].str.strip()

#Display
data.head()
```
